```python
import xgboost as xgb
import numpy as np
import mltools as ml

X = np.genfromtxt('HW4-code\X_train.txt', delimiter=None)
Y = np.genfromtxt('HW4-code\Y_train.txt', delimiter=None)
Xte = np.genfromtxt('HW4-code\X_test.txt', delimiter=None)
wts = np.genfromtxt('HW4-code\weights.txt', delimiter=None)

X = np.hstack((X,X.dot(wts)))
Xte = np.hstack((Xte,Xte.dot(wts)))

X,Y = ml.shuffleData(X,Y)

Xtr, Xva, Ytr, Yva = ml.splitData(X, Y)
X, params = ml.rescale(X)

Xt, Yt = Xtr, Ytr
Xv, Yv = Xva, Yva
```

```python
dtrain = xgb.DMatrix(Xt, label=Yt)
dval = xgb.DMatrix(Xv, label=Yv)
dtest = xgb.DMatrix(Xte)
evallist  = [(dval,'eval'), (dtrain,'train')]
num_round = 1000

eta = [0.01,0.015,0.025,0.05,0.1]
gamma = [0.05,0.1,0.3,0.5,0.7,0.9,1.0]
maxDepth = [1,2,3,5,7,9]
min_child_weight = [1,3,5,7]
subsample = [0.6, 0.7, 0.8, 0.9, 1.0]
colsample_bytree = [0.6, 0.7, 0.8, 0.9, 1.0]
```

```python
param = {'bst:max_depth':3, 'bst:eta':0.1, 'silent':1, 'objective':'binary:logistic' }
param['nthread'] = 4
param['eval_metric'] = 'auc'
bst = xgb.train(param, dtrain, num_round, evallist)
bst.dump_model('dump.raw.txt')
```

```python
ypred = bst.predict(dtest)
Yte = np.vstack((np.arange(Xte.shape[0]), ypred)).T
np.savetxt('Y_submit_xgboosted.txt', Yte, '%d, %.2f', header='ID,Prob1', comments='', delimiter=',')
```
